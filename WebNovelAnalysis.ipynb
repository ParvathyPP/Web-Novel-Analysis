```python
import io
import nltk
from nltk.tokenize import RegexpTokenizer
from collections import Counter
count = 1
total_para = 0
tot_words = []
first_half_words = []
second_half_words = []
tokenizer = RegexpTokenizer('\w+')
sw = nltk.corpus.stopwords.words('english')
data = []
for i in links:
    d =[]
    words = []
    total_word_len = 0
    total_word_count = 0
    word_count = 0
    url = str(i)
    #Make the request
    page = requests.get(url)
    #Extract HTML from Response object and Create a BeautifulSoup object from the HTML
    soup = BeautifulSoup(page.text, 'html5')
    #get all p tag from it
    elems = soup.find_all('p')
    total_para += len(elems)
    print(len(elems))
    file_name = "chapter_"+ str(count)+".txt"
    my_file = io.open(file_name, 'w', encoding = "utf-8")
    print(file_name)
    for job_elem in elems:
        s = BeautifulSoup(str(job_elem))
        pattern = re.compile('<.*?>')
        base_text = re.sub(pattern, '', str(job_elem))
        my_file.write(base_text)
        my_file.write("\n\n")
        tokens = tokenizer.tokenize(base_text)
        for word in tokens:
            if word.lower() not in sw:
                words.append(word.lower())
                tot_words.append(word.lower())
                total_word_len += len(word)
                word_count += 1
                l = (len(links) / 2)
                if count <= l:
                    first_half_words.append(word.lower())
                else:
                    second_half_words.append(word.lower())
        #print(base_text)
        #print(words)
    count+=1
    counters_found = Counter(words)
    most_occur = counters_found.most_common(5)
    total_words_used = len(words)
    d.append(total_words_used)
    d.append(len(elems))
    average_word_len = total_word_len / word_count
    d.append(average_word_len)
    d.append(word_count / len(elems))
    d.append(most_occur)
    print(d)
    data.append(d)
print(data)
#myfile.close()
#text_file.close()
```

```python
import pandas as pd
columns = ['total_words_used', 'total_paragraphs', 'average_word_length', 'average_words_per_paragraphs', 'most_frequent_words']
train = pd.DataFrame(data, columns=columns)
```

```python
from nltk import word_tokenize
tokens = word_tokenize(novel)
```

```python
from nltk import pos_tag, word_tokenize
from nltk.corpus import words
proper_nouns = []
i = 0
tags = pos_tag(tokens)
while i < len(tags):
    #if tags[i][1] not in words.words():
    if tags[i][1] == 'NNP':
        if tags[i+1][1] == 'NNP':
            proper_nouns.append(tags[i][0].lower() + " " + tags[i+1][0].lower())
            i+= 1
        else:
            proper_nouns.append(tags[i][0].lower())
    i+=1
```
